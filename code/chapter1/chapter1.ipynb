```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl
```

```python
s=ctl.tf("s")
P = 1/(10*s**2+0.1*s)
P
```

```python
C = 20000*(s+0.1)/(s+100)
C
```

```python
T = C*P/(1+C*P)
tvec = np.arange(0,1,0.001)
tv,y = ctl.step_response(T,tvec)
```

```python
r = np.ones(tvec.shape)
plt.figure(figsize=(4,3))
plt.plot(tvec,r,'b-.',tvec,y,'r-')
plt.xlabel('time (sec)')
plt.ylabel('y(t)')
plt.legend(['step ref.','plant output'])
plt.show()
```
